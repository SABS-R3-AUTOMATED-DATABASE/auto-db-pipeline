```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def extract_url_without_patents(source):
    urls = []
    for _ in source:
        if len(_.split("(h")) > 1:
            url = _.split("(")[1]
            url = url.split(")")[0]
            if url not in urls and len(url) != 0 and 'patent' not in url:
                urls.append(url)
    return urls

def extract_url(source):
    urls = []
    for _ in source:
        if len(_.split("(h")) > 1:
            url = _.split("(")[1]
            url = url.split(")")[0]
            if url not in urls and len(url) != 0:
                urls.append(url)
    return urls
df = pd.read_csv (r'covabdab.csv')
print(len(extract_url(df["Sources"])))
list_of_urls = extract_url_without_patents(df["Sources"])
with open("url.txt", "w") as f:
        for name in list_of_urls:
            f.write(str(name) + "\n")
            
```

```python
import pandas as pd
df = pd.read_csv('covabdab.csv')
papers = []
rawlist = []
urllist = []
for x in set(df['Sources']):
    papers.append(x)
for s in papers:
    rawlist.append(s[s.find("(")+1:s.find(")")])
for _ in rawlist:
    if _ not in urllist and 'patent' not in _ and 'rcsb' not in _:
        urllist.append(_)
        
with open("urls.txt", "w") as f:
        for name in urllist:
            f.write(str(name) + "\n")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
def get_title_from_url(list):
    results = []
    for _ in list:
        print(_)
        headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2919.83 Safari/537.36'}
        r = requests.get(_ , headers=headers)
        soup = BeautifulSoup(r.content, 'html.parser')
        if soup.title != None:
            results.append(soup.title.string)
            print(soup.title.string)
        else:
            results.append('na')
            print('na')
    return results
with open('urls.txt') as f:
    urls = f.read().splitlines()
titles = get_title_from_url(urls)
```

```python

title = []
for t in titles:
    t = t.split(" - S")[0]
    t = t.split(" | ")[0]
    t = t.split(": C")[0]
    title.append(t)
with open("titles_edited1.txt", "w") as f:
        for name in title:
            f.write(str(name) + "." + "\n")
```
