```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


def extract_url_without_patents(source):
    urls = []
    for _ in source:
        if len(_.split("(h")) > 1:
            url = _.split("(")[1]
            url = url.split(")")[0]
            url = url.split('.pdf')[0]
            if url not in urls and len(url) != 0 and 'patent' not in url and 'rcsb' not in url and url != 'https://www.cell.com/action/showPdf?pii=S1074-7613%2821%2900359-9':
                urls.append(url)
    return urls

def get_title_from_url(list):
    results = []
    for _ in list:
        print(_)
        headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2919.83 Safari/537.36'}
        r = requests.get(_ , headers=headers)
        soup = BeautifulSoup(r.content, 'html.parser')
        if soup.title != None:
            results.append(soup.title.string)
    return results

df = pd.read_csv (r'covabdab.csv')
list_of_urls = extract_url_without_patents(df["Sources"])
titles = get_title_from_url(list_of_urls)
with open("titles.txt", "w") as f:
        for name in titles:
            f.write(str(name) + "\n")
with open("urls_raw.txt", "w") as f:
        for name in urls:
            f.write(str(name) + "\n")
```

```python
with open('titles.txt') as f:
    titles = f.read().splitlines()
    
title = []
for t in titles:
    t = t.split(" - S")[0]
    t = t.split(" | ")[0]
    title.append(t)
with open("titles_edited.txt", "w") as f:
        for name in title:
            f.write(str(name) + "." + "\n")
```

```python
with open('titles_edited.txt') as f:
    titles_covab = f.read().splitlines()
    
with open('search_results.txt') as f:
    titles_search = f.read().splitlines()

titles_covab_l = [elem.lower() for elem in titles_covab]   
titles_search_l = [elem.lower() for elem in titles_search] 
not_searched = [] 
yes = 0
for elem in titles_covab_l:
    if elem.replace('sars-cov-2','sars-cov-2') not in titles_search_l:
        not_searched.append(elem)
    else:
        yes += 1
print(yes)
print(not_searched)
```
